```python
import numpy as np
import pandas as pd
import sklearn
import os
import glob
```

## Data load

```python
path = os.getcwd()
csv_files = glob.glob(os.path.join(path, "*.csv"))
# loop over the list of csv files
for f in csv_files:

    # read the csv file
    df = pd.read_csv(f)

    # print the location and filename
    print('Location:', f)
    print('File Name:', f.split("\\")[-1])

    print(df.columns)
    # print the content
    print('Content:')
    print(df.head())

    print('-' * 80)
```
